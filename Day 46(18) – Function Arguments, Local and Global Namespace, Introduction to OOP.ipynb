```python
# Day 46 - Bank Account Manager
# Topics: Function Arguments, Local/Global Namespace, OOP

# Global variable (transaction counter)
transaction_count = 0

class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        global transaction_count
        if amount > 0:
            self.balance += amount
            transaction_count += 1
            print(f"üí∞ Deposited: {amount}")
        else:
            print("‚ùå Invalid deposit amount")

    def withdraw(self, amount):
        global transaction_count
        if amount <= 0:
            print("‚ùå Invalid withdrawal amount")
        elif amount > self.balance:
            print("‚ùå Insufficient balance! Withdrawal denied.")
        else:
            self.balance -= amount
            transaction_count += 1
            print(f"üí∏ Withdrawn: {amount}")

    def display_balance(self):
        print(f"üë§ Account Holder: {self.account_holder}")
        print(f"üíµ Current Balance: {self.balance}")


print("=== Bank Account Manager ===")

# object creation using positional args
account1 = BankAccount("Arun", 5000)

# method calls with positional arguments
account1.deposit(2000)
account1.withdraw(1500)
account1.display_balance()

print("\n--- Using Keyword Arguments ---")
# object creation using keyword args
account2 = BankAccount(account_holder="Priya", balance=3000)

# method calls with keyword arguments
account2.deposit(amount=2500)
account2.withdraw(amount=1000)
account2.display_balance()

# Display total number of transactions (global variable used)
print(f"\nüî¢ Total Transactions Processed: {transaction_count}")
```
