```python
# Day 45 - Simple Calculator App
# Topics: Built-in Functions + User-Defined Functions

# ---------- User-defined functions ----------
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "❌ Division by zero is not allowed!"
    return a / b


print("=== Simple Calculator App ===")

# ---------- Input ----------
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
except ValueError:
    print("❌ Invalid input! Numbers only.")
    raise SystemExit()

operation = input("Enter operation (+, -, *, /): ")

# ---------- Perform calculation ----------
if operation == "+":
    result = add(num1, num2)
elif operation == "-":
    result = sub(num1, num2)
elif operation == "*":
    result = mul(num1, num2)
elif operation == "/":
    result = div(num1, num2)
else:
    print("❌ Invalid operation!")
    raise SystemExit()

print("\nResult:", result)

# ---------- Using built-in functions ----------
print("\n--- Built-in Functions Demonstration ---")
print("abs(result):", abs(result) if isinstance(result, (int, float)) else result)
print("round(result, 2):", round(result, 2) if isinstance(result, (int, float)) else result)
print("pow(num1, num2):", pow(num1, num2))
```
