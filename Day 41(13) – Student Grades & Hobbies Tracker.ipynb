```python
# Day 41 - Dictionary, Set, Frozenset
# Project: Student Grades & Hobbies Tracker

# Dictionary -> student name : grade
grades = {}

print("Enter / Update grades for 3 students:")
for i in range(3):
    name = input(f"Student {i+1} Name: ")
    grade = input("Grade: ")
    grades[name] = grade   # add/update
print("\nCurrent Grades:")
print(grades)


# Set -> store hobbies (no duplicates)
hobbies = set()
print("\nEnter 3 hobbies:")
for i in range(3):
    hobby = input(f"Hobby {i+1}: ")
    hobbies.add(hobby)

print("\nHobbies (Set):", hobbies)


# Convert set to frozenset (Immutable)
hobby_frozen = frozenset(hobbies)
print("\nHobbies as Frozenset (Immutable):", hobby_frozen)
```
