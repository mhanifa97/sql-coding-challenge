```python
# Day 44 - Exam Scores Report
# Topics: range, for loop, enumerate, list comprehension

import random

# 1️⃣ Create a list of 10 random exam scores (0–100) using list comprehension
scores = [random.randint(0, 100) for _ in range(10)]
print("Exam Scores:", scores)

# 2️⃣ Display each score with its index using enumerate
print("\n--- Scores with Index ---")
for index, value in enumerate(scores):
    print(f"Score {index}: {value}")

# 3️⃣ Find and display all scores > 75
high_scores = [score for score in scores if score > 75]
print("\nScores greater than 75:", high_scores)
```
