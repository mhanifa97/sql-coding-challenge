```python
# Day 47 - Student Management System
# Topics: Dictionary, List, Functions, Tuples/Frozenset, Sorting, Exception Handling

# Constant Data (cannot be changed)
available_classes = ("Class A", "Class B", "Class C")

# Students stored as List of Dictionaries
students = []   # each record: {"name": "", "age": 0, "grade": 0, "class": ""}


# ---- Functions ----

def add_student():
    try:
        name = input("Enter Student Name: ")
        age = int(input("Enter Age: "))
        grade = float(input("Enter Grade/Marks: "))
        print("Available Classes:", available_classes)
        class_name = input("Enter Class: ")

        if class_name not in available_classes:
            print("‚ùå Invalid class name!")
            return

        students.append({"name": name, "age": age, "grade": grade, "class": class_name})
        print("üëç Student Added Successfully!")
    except ValueError:
        print("‚ùå Invalid input! Age/Grade must be a number.")


def delete_student():
    name = input("Enter Student Name to Delete: ")
    for s in students:
        if s["name"].lower() == name.lower():
            students.remove(s)
            print("üóë Student Deleted!")
            return
    print("‚ùå Student not found!")


def update_student():
    name = input("Enter Student Name to Update: ")
    for s in students:
        if s["name"].lower() == name.lower():
            try:
                s["age"] = int(input("Enter New Age: "))
                s["grade"] = float(input("Enter New Grade: "))
                print("Available Classes:", available_classes)
                class_name = input("Enter New Class: ")
                if class_name in available_classes:
                    s["class"] = class_name
                    print("üîÑ Student Updated!")
                else:
                    print("‚ùå Invalid class name!")
                return
            except ValueError:
                print("‚ùå Invalid input! Age/Grade must be numbers.")
                return
    print("‚ùå Student not found!")


def search_student():
    name = input("Enter Student Name to Search: ")
    found = False
    for s in students:
        if name.lower() in s["name"].lower():
            print(s)
            found = True
    if not found:
        print("‚ùå Student not found!")


def performance_report():
    print("\n--- Performance Category ---")
    for s in students:
        if s["grade"] >= 90:
            category = "üèÜ Excellent"
        elif s["grade"] >= 75:
            category = "üëç Good"
        elif s["grade"] >= 50:
            category = "üôÇ Average"
        else:
            category = "‚ö†Ô∏è Poor"
        print(f"{s['name']} -> {category}")


def sort_students():
    students.sort(key=lambda x: x["name"])
    print("üìå Students Sorted by Name!")


def display_all():
    if not students:
        print("‚ö†Ô∏è No Students Found")
        return
    for s in students:
        print(s)


# ---- Main Menu ----
while True:
    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. Delete Student")
    print("3. Update Student")
    print("4. Search Student")
    print("5. Show Performance Report")
    print("6. Sort Students by Name")
    print("7. Display All Students")
    print("8. Exit")

    try:
        choice = int(input("Enter choice: "))
    except ValueError:
        print("‚ùå Invalid input! Enter a number.")
        continue

    if choice == 1:
        add_student()
    elif choice == 2:
        delete_student()
    elif choice == 3:
        update_student()
    elif choice == 4:
        search_student()
    elif choice == 5:
        performance_report()
    elif choice == 6:
        sort_students()
    elif choice == 7:
        display_all()
    elif choice == 8:
        print("üëã Exiting Program‚Ä¶ Bye!")
        break
    else:
        print("‚ùå Invalid Choice!")
```
