```python
# Day 1 - NumPy Arrays & Operations

import numpy as np

print("=== NumPy Arrays & Operations ===")

# 1️⃣ Create 1D, 2D, 3D arrays
arr_1d = np.array([10, 20, 30, 40])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print("1D Array:", arr_1d)
print("2D Array:\n", arr_2d)
print("3D Array:\n", arr_3d)

# 2️⃣ Arithmetic operations
print("\n--- Arithmetic Operations ---")
print("Addition:", arr_1d + 5)
print("Subtraction:", arr_1d - 5)
print("Multiplication:", arr_1d * 2)
print("Division:", arr_1d / 2)

# 3️⃣ Statistical functions
print("\n--- NumPy Functions ---")
print("Mean:", np.mean(arr_1d))
print("Sum:", np.sum(arr_1d))
print("Standard Deviation:", np.std(arr_1d))

# 4️⃣ Reshape, arange, linspace
reshaped_arr = np.arange(1, 13).reshape(3, 4)
print("\nReshape using arange:\n", reshaped_arr)

lin_arr = np.linspace(0, 1, 5)
print("Linspace (0 to 1):", lin_arr)

# 5️⃣ Index of max/min values
print("\nMax value index:", np.argmax(arr_1d))
print("Min value index:", np.argmin(arr_1d))
```
