```python
# Day 40 - Contact Directory
# Topic: List Methods and Slicing

# 2D List -> Each row: [Name, Phone, Email]
contacts = [
    ["Arun", "9876543210", "arun@gmail.com"],
    ["Priya", "9123456789", "priya@yahoo.com"],
    ["Kumar", "9000012345", "kumar@outlook.com"]
]

print("Initial Contacts:")
for c in contacts:
    print(c)

# Add new contact
print("\n--- Add New Contact ---")
name = input("Enter Name: ")
phone = input("Enter Phone: ")
email = input("Enter Email: ")
contacts.append([name, phone, email])
print("New contact added successfully!\n")

# Sort contacts alphabetically (Aâ€“Z) by name
contacts.sort(key=lambda x: x[0])

print("--- Sorted Contacts (A-Z) ---")
for c in contacts:
    print(c)

# Search contact by partial name match
print("\n--- Search Contact ---")
key = input("Enter name to search: ").lower()
print("\nSearch Results:")
found = False
for c in contacts:
    if key in c[0].lower():   # partial match
        print(c)
        found = True

if not found:
    print("No matching contacts found.")

# Display only first 3 contacts (Slicing)
print("\n--- Top 3 Contacts After Sorting ---")
for c in contacts[:3]:
    print(c)
```
