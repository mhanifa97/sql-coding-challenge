```python
# Day 43 - Number Guessing Game
# Topic: while loop, else, break, continue

import random

print("=== Number Guessing Game ===")
secret_number = random.randint(1, 20)
attempts = 0
max_attempts = 5

while attempts < max_attempts:
    try:
        guess = int(input(f"Attempt {attempts+1}/{max_attempts} - Guess a number between 1 and 20: "))
    except ValueError:
        print("âŒ Invalid input! Enter only numbers.")
        continue  # don't count invalid attempts

    if guess < 1 or guess > 20:
        print("âš ï¸ Guess must be between 1 and 20.")
        continue

    attempts += 1

    if guess == secret_number:
        print(f"ğŸ‰ Correct! You guessed the number in {attempts} attempts.")
        break
    elif guess < secret_number:
        print("ğŸ“‰ Too low! Try again.")
    else:
        print("ğŸ“ˆ Too high! Try again.")

else:
    # else executes only if loop ends without break
    print(f"âŒ You failed to guess in {max_attempts} attempts.")
    print(f"ğŸ’¡ The correct number was: {secret_number}")
```
