```python
# Day 39 - Tuple, Immutable vs Mutable, List Creation
# Project: Retail Inventory Manager

# Tuple -> Fixed categories (Immutable)
product_categories = ("Electronics", "Stationery", "Clothing")

# List -> Dynamic products (Mutable)
products = ["Mobile", "Pen", "Shirt"]

print("Initial Product Categories:", product_categories)
print("Initial Product List:", products)

# Add new product
new_product = input("Enter a new product to add: ")
products.append(new_product)
print(f"Added: {new_product}")

# Remove product by index
print("\nProducts with index:")
for idx, item in enumerate(products):
    print(f"{idx}: {item}")

remove_index = int(input("Enter the index of product to remove: "))
removed_item = products.pop(remove_index)
print(f"Removed: {removed_item}")

# Final output
print("\nFinal Product Categories (Tuple):", product_categories)
print("Final Product List (List):", products)
```
