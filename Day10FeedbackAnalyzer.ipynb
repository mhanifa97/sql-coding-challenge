```python
# Day 38 – Feedback Analyzer

# 1. Ask the user to input feedback
feedback = input("Enter your feedback: ")

# 2. Count how many words are in the sentence
word_count = len(feedback.split())

# 3. Count how many times the word “data” appears (case-insensitive)
data_count = feedback.lower().count("data")

# 4. Extract first 5 and last 5 characters
first_5 = feedback[:5]
last_5 = feedback[-5:]

# 5. Reverse the sentence
reversed_sentence = feedback[::-1]

# ----------- OUTPUT ---------------
print("\n----- Feedback Analysis Report -----")
print("Total words            :", word_count)
print("Occurrences of 'data'  :", data_count)
print("First 5 characters     :", first_5)
print("Last 5 characters      :", last_5)
print("Reversed sentence      :", reversed_sentence)
```
